```python
import pandas as pd
import numpy as np
```

```python
#Import csv files

data201904 = "Resources/Citibike/JC-201904-citibike-tripdata.csv"
data201905 = "Resources/Citibike/JC-201905-citibike-tripdata.csv"
data201906 = "Resources/Citibike/JC-201906-citibike-tripdata.csv"
data201907 = "Resources/Citibike/JC-201907-citibike-tripdata.csv"
data201908 = "Resources/Citibike/JC-201908-citibike-tripdata.csv"
data201909 = "Resources/Citibike/JC-201909-citibike-tripdata.csv"


citibike201904 = pd.read_csv(data201904)
citibike201905 = pd.read_csv(data201905)
citibike201906 = pd.read_csv(data201906)
citibike201907 = pd.read_csv(data201907)
citibike201908 = pd.read_csv(data201908)
citibike201909 = pd.read_csv(data201909)


tripdata201904_df = citibike201904
tripdata201905_df = citibike201905
tripdata201906_df = citibike201906
tripdata201907_df = citibike201907
tripdata201908_df = citibike201908
tripdata201909_df = citibike201909


tripdata201912_df.head(5)
```

```python
#Combine the CSV Files into a dataframe
trip_df = [tripdata201904_df,tripdata201905_df,tripdata201906_df,
           tripdata201907_df,tripdata201908_df,tripdata201909_df,]
completetripdataresults = pd.concat(trip_df)
completetripdataresults
```

```python
completetripdataresults.dtypes
```

```python
completetripdataresults['gender'] = completetripdataresults['gender'].astype(str)
completetripdataresults.dtypes
```

```python
update_gender = completetripdataresults['gender']
update_male = update_gender.replace('1', 'Male')
update_female = update_male.replace('2', 'Female')
update_unkown = update_female.replace('0', 'Unknown')
completetripdataresults['gender'] = update_unkown
completetripdataresults
```

```python
completetripdataresults.count()
```

```python
completetripdataresults=completetripdataresults.dropna(how="any")
completetripdataresults.count()
```
